# Merging DataFrames
#
# inner (defualt), outer(i.e full), entire left, entire right 

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"D:\My Projects 1\Dataanalytics (begining)\Beginingdata files\LOTR.csv")
df2 = pd.read_csv(r"D:\My Projects 1\Dataanalytics (begining)\Beginingdata files\LOTR 2.csv")
df1
```

```python
df2
```

# merge method   
#

# inner join parameters

```python
df1.merge(df2,)
```

```python
df1.merge(df2, how='inner', on = ['FellowshipID'], suffixes = ('_df1','_df2'))

# do shift tab to see more parameters
```

# 
# outer(full) join 

```python
df1.merge(df2, how = 'outer')
```

#
# right join  (didnt add id 1004,1003)

```python
df1.merge(df2, how = 'right')
```

# left join (similary only added the common between the two df) 

```python
# print(df1)
df1.merge(df2, how = 'left')
```

```python
# cross join(not widely used) 
# cross join (each value of 1st df being merged with will be compared to the other df)
```

```python
df1.merge(df2, how = 'cross')
```

#
# Join methods, (better when working with index)
#
# similar to merge but more confusing
# it doesnt provide us with default lsuffix and rsuffix(can check by doing shift tab)

# 
# parameter doesnt works as intended

```python
df1.join(df2, lsuffix= '_df1', rsuffix = '_df2')
```

```python
df1.join(df2,  on =['FellowshipID'], how = 'outer', lsuffix= '_df1', rsuffix = '_df2')
```

#
# Concat (instead of 'how' ithas 'join') 
# 
# Simply connect multiple df accoridng to the axis

```python
# its inner join just finds the common column of both
pd.concat([df1,df2], join = 'inner')
```

```python
# connect new df downward by doing axis = 0(default)
pd.concat([df1,df2], join = 'outer')
```

```python
# connect new df downward by doing axis = 0(default)
pd.concat([df1,df2], join = 'outer', axis = 1)
```
