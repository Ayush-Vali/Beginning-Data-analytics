```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\My Projects 1\Dataanalytics (begining)\Beginingdata files\Ice Cream Ratings.csv")
df.set_index('Date', inplace = True)
df
```

```python
df.plot() # default kind = 'Line' 
```

# can specify kind of plot outside

```python
df.plot.hist()
```

# Can do specific column plot

```python
df['Overall Rating'].plot()
```

# For bar, (stacked parameter unique and has barhorizontal)

```python
df.plot.barh()
```

```python
# stacked normal

df.plot(kind = 'bar', title = 'Ice Cream Rating', xlabel ='Daily readings', ylabel = 'Scores', stacked=True)
```

```python
# same as above
df.plot(kind = 'barh', title = 'Ice Cream Rating', xlabel ='Daily readings', ylabel = 'Scores', stacked=True)
```
