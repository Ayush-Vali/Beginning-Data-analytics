# replace
#
# in it caret(^) represents we aren't going to return any char except the ones specified, with whatever the next char

```python
# in this example, after ^small and capital a to z, as well as numbers, are only allowed,
# others are replaced with specified "nothing i.e ,'' " 
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','', regex = True)
df
```

# Now to convert all the phone number in this format
# Used: apply method, lambda expression, slicing,
# apply method iterate over all elements of column

```python
# df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' )  """ Gives int error, so convert to string

# df['Phone_Number'] = str(df['Phone_Number'])  #"""" THis wont work as intended
# THEREFORE
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))  # now each phone number is turned to string

type(df['Phone_Number']) # check converted or not

# now can use this
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

df
```

#
# Now to get rid of -- in front of nan, using replace

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
df
```

# Need to give list of contactable people and drop other rows 

```python
# remove from  Do_Not_Contact column, the people who doesnt want to be contacted 
```

```python

# Logic is to loop through the df.index which gives list of indexes, do print(df.index)&print(df['name']) and check 

for x in df.index:
    if df.loc[x, 'Do_Not_Contact' ] == 'Y':  # x is row, 'donotcontact' column which should be only called in that row
        df.drop(x, inplace = True)

df
```
